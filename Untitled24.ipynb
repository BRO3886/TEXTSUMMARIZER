```python
from urllib import request 
from bs4 import BeautifulSoup as bs
import re
import nltk
import heapq
```

```python
url= "https://en.wikipedia.org/wiki/Blog"
```

```python
doc=request.urlopen(url)
```

```python
soupobj=bs(doc,'html.parser')
```

```python
 paracontents=soupobj.findAll('p')
```

```python
print( paracontents)
```

```python
allparacontent=""
```

```python
for  paracontent in  paracontents:
    allparacontent+= paracontent.text
```

```python
allparacontent_cleanedDataa=re.sub(r'\[[0-9]*\]',' ',allparacontent)
 
```

```python
allparacontent_cleanedData=re.sub(r'\s+',' ',allparacontent)
```

```python
 print(allparacontent_cleanedData)
```

```python
print(allparacontent_cleanedDataa)
```

```python
sentence_tokens=nltk.sent_tokenize(allparacontent_cleanedDataa)
word_tokens=nltk.word_tokenize(allparacontent_cleanedDataa)
```

```python
stopwords=nltk.corpus.stopwords.words('english')
```

```python
word_frequencies={}
```

```python
for word in word_tokens:
    if word not in stopwords:
        if word not in word_frequencies.keys():
            word_frequencies[word]=1
        else:
            word_frequencies[word]+=1
print(word_frequencies)
```

```python
maximum_frequency_word=max(word_frequencies.values())
print(maximum_frequency_word)
```

```python
for word in word_frequencies.keys():
    word_frequencies[word]=(word_frequencies[word]/maximum_frequency_word)
print(word_frequencies)
```

```python
sentence_score={}
for sentence in sentence_tokens:
    for word in nltk.word_tokenize(sentence.lower()):
        if word in word_frequencies.keys():
            if (len(sentence.split(' ')))<30:
                if sentence not in sentence_score.keys():
                    sentence_score[sentence]=word_frequencies[word]
                else:
                    sentence_score[sentence]+=word_frequencies[word]
print(sentence_score)
     
```

```python
summary=heapq.nlargest(10,sentence_score,key=sentence_score.get)
print(summary)
```
